```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('yahoo_stock.csv')

df['SMA'] = df['Close'].rolling(window = 100).mean()
df['EMA'] = df['Close'].rolling(window = 20).mean()
df_fill = df['SMA'].mean()
df['SMA'] = df['SMA'].fillna(df_fill)
```

```python
plt.plot(df['Date'],df['SMA'])
plt.plot(df['Date'],df['Close'],color ='Green')
plt.plot(df['Date'],df['EMA'],color = 'red')
```

```python
df['DC'] = df['Close'].diff()
plt.plot(df['Date'],df['DC'])
```
